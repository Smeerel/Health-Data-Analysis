```python
def uniplot(df,col,title,hue =None):

    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30


    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright')

    plt.show()
```

# Conclusion:
## Ineffectiveness of the SMS reminder service is evident in the following observations:

- A significant portion of patients does not receive SMS reminders for their upcoming appointments.

- Although a good proportion of young adults recieve SMS reminders but a relatively low prevalence of chronic illnesses, such as Diabetes and Hypertension among this age group, makes them more prone to missing appointments.

- Among patients who consume alcohol, males have a slightly higher likelihood of missing appointments, and they are the least likely to receive SMS notifications.

- Regardless of their medical condition and disability status, the majority of patients do not receive SMS reminders about their upcoming appointments.

- Patients enrolled in scholarship programs also face challenges with SMS reminders, as a higher percentage of them do not receive notifications and have an increased risk of not showing up.


## Gender's Role:

Females, across all age groups, exhibit a higher propensity to miss their appointments, even when they have:

- Chronic medical conditions.
- Enrollment in scholarship programs.
- Received SMS reminders.


## Highly Neglected Groups:

The most neglected groups in terms of appointment attendance include pediatric,very elderly patients and handicapped patients, with some common reasons being:

- Lack of SMS reminders for their appointments.

- Low prevalence of chronic conditions in pediatric patients.

- Higher chances of appointment non-attendance when the disability level is 3 or more. This level of disability is frequently associated with pediatric (1-17) and older age(69-85) groups.


## Regular Patients:

- A significant majority (75%) of patients who consistently attend their appointments are above the age of 50.
- Conditions like Hypertension and Diabetes are more prevalent among seniors, making patients in the age group of 50-100 years less likely to miss appointments as they need active medical attention and followups.
